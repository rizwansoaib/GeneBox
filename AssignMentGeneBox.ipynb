## Problem Statement:- 

### Write a program to find the indices of two numbers in a list that add upto a given number. (Try to find an O(n) algorithm)

```python
'''
Solution of problem will be we can use dictionary where key will be number and value will be index. We travers dictionary then for each value if MyNumber-CurrentNumber in Mydict
then we returning those indexes.

'''

def dictionary_solution(mylist,mynumber):
    mydict={}
    for index in range(len(mylist)):
        mydict[mylist[index]]=index
        
    Answerlist=[]
        
    for key,value in mydict.items():
        if mydict.get(mynumber-key,None):
            if mydict[key]<mydict[mynumber-key]: #for remove duplicate and same index like 2,4==4,2 or 2,2
                Answerlist.append((mydict[key],mydict[mynumber-key]))
    
    return Answerlist


'''
Solution from scratch using hashing this code will work only for postive numbers only but we can extend 
it for negative numbers also for that we will do scaling find first minimum of list then add into all numbers
of list.

'''



def AnotherSolution(mylist,mynumber):
    max_numbers=max(mylist)
    number_hash=[(False,None)]*(max_numbers+1)
    
    for index in range(len(mylist)):
        number_hash[mylist[index]]=(True,index)
        
    Answerlist=[]    
    for i in range(mynumber//2+1):
        start=i
        remain=mynumber-i
        if start>max_numbers:
            break
        if remain>max_numbers:
            continue
        if number_hash[start][0] and number_hash[remain][0]:
            smaller_index=min(number_hash[i][1],number_hash[mynumber-i][1])
            greater_index=max(number_hash[i][1],number_hash[mynumber-i][1])
            
            if smaller_index!=greater_index: #for not same index 
                Answerlist.append((smaller_index,greater_index))
        
    return Answerlist
    
            
if __name__=='__main__':     
    
    #mylist=list(map(int,input('Please enter list numbers separted by space: ').split()))
    #mynumber=int(input('Enter given number: '))
    
    mylist=[10,9,2,3,4,11,7,8,6]
    mynumber=13
    



        
    print(mylist,mynumber)    
    
    Answerlist=dictionary_solution(mylist,mynumber)
    print('\nFrom Dictionary Solution:')
    if Answerlist:
        for first_index,second_index in Answerlist:
            print('indexes are {} and {} value are {}+{}={}'.format(first_index,second_index,mylist[first_index],mylist[second_index],mynumber))
    else:
        print('no index found')
        
    print('\nFrom Scratch Solution:')    
    Answerlist=AnotherSolution(mylist,mynumber)

    if Answerlist:
        for first_index,second_index in Answerlist:
            print('indexes are {} and {} value are {}+{}={}'.format(first_index,second_index,mylist[first_index],mylist[second_index],mynumber))
    else:
        print('no index found')
```

## Problem 2(b):
### Replace the first column in each row with the species name. For the given example, Bifidobacterium_pseudocatenulatum.

```python
df=dataframe[['#SampleID']]


for index in range(len(df)):
    if len(df.iloc[index]['#SampleID'].split('s__'))>1:
        species_name=df.iloc[index]['#SampleID'].split('s__')[1].split('|')[0]
        df.iloc[index]['#SampleID']=species_name
        
           
dataframe['#SampleID']=df['#SampleID']
dataframe.to_csv('solution2.csv')
dataframe
```
